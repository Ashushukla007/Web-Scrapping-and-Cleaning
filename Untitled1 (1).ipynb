```python
import pandas as pd
import numpy as np
```

```python
d=pd.read_csv('Companies_growth.csv')
```

```python
d['EV M/B']=d['EstimatedValuation'].str[-1]
d['TFR M/B']=d['Total FundingRaised'].str[-1]
d['LAR M/B']=d['Latest AmountRaised'].str[-1]

d['EstimatedValuation']=d['EstimatedValuation'].str[:-1]
d['Total FundingRaised']=d['Total FundingRaised'].str[:-1]
d['Latest AmountRaised']=d['Latest AmountRaised'].str[:-1]
```

```python
d.sample(50)
```

```python
d['EstimatedValuation']=d['EstimatedValuation'].replace('-',np.nan)
d['Total FundingRaised']=d['Total FundingRaised'].replace('-',np.nan)
d['Latest AmountRaised']=d['Latest AmountRaised'].replace('457,44',"45744")
d['Latest AmountRaised']=d['Latest AmountRaised'].replace('-',np.nan)
d['Latest AmountRaised']=d['Latest AmountRaised'].str.split(',').str.join(',')
     
```

```python
d['Latest AmountRaised']=d['Latest AmountRaised'].str.split(',').str.join(',')
     
d=d.astype({'EstimatedValuation':'float',
          'Total FundingRaised' :'float',
           'Latest AmountRaised':'float'
           })
```

```python
d
```

```python
d.dtypes
```

```python
if d['EV M/B'].str=='M':
    d['EstimatedValuation']=d['EstimatedValuation']*1000000
d['EstimatedValuation']=d['EstimatedValuation']*1000000000
if d['TFR M/B'].str=='M':
    d['Total FundingRaised']=d['Total FundingRaised']*1000000
d['Total FundingRaised']=d['Total FundingRaised']*1000000000
if d['LAR M/B'].str=='M':
    d['Latest AmountRaised']=d['Latest AmountRaised']*1000000
d['Latest AmountRaised']=d['Latest AmountRaised']*1000000000
```

```python
d=d.drop(columns=['LAR M/B','EV M/B','TFR M/B'])
```

```python
d
```

```python
import datetime as dt
```

```python
d['Latest FundingDate']=pd.to_datetime(d['Latest FundingDate'])
```

```python
d.dtypes
d['Year']=d['Latest FundingDate'].dt.year
d['Month']=d['Latest FundingDate'].dt.month_name()
d['Date']=d['Latest FundingDate'].dt.day
d['Weekday']=d['Latest FundingDate'].dt.day_name()
```
